Description: Analyze monthly sales data for multiple products, computing statistics and visualizing trends.

Steps:
1. Load the CSV using pandas.read_csv() and convert to NumPy arrays.
2. Compute mean, median, and standard deviation for each product’s sales.
3. Identify the top-performing product (highest average sales).
4. Use Matplotlib to plot sales trends over months.

```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('sales_data.csv')
```

```python
print(df.head())
```

```python
print(f'Mean : {round(df['Sales'].mean(),2)}')
print(f'Median : {round(df['Sales'].median(),2)}')
print(f'Standard Deviation : {round(df['Sales'].std(),2)}')
```

```python
# Identify the top-performing product
print(f'Top Performing Product : {df.groupby("Product")["Sales"].mean().idxmax()}')
```

```python
# Use Matplotlib to plot sales trends over months
import matplotlib.pyplot as plt

df.groupby('Month')['Sales'].sum().plot(kind='line')
plt.title('Sales Trends Over Months')
plt.xlabel('Month')
plt.ylabel('Total Sales')

plt.show()
```

```python

plt.savefig('sales_trends.png')
df.to_csv('sales_analysis_results.csv', index=False)
```
